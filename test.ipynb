```python
import pandas as pd
import duckdb
import creds

conn = duckdb.connect("my_db.duckdb")
conn.execute("INSTALL mysql;")
conn.execute(f"ATTACH 'host={creds.host} user={creds.user} password={creds.password} port={creds.port} database={creds.database}' AS mysqldb (TYPE mysql)")
```

```python
conn.execute("USE mysqldb")
```

```python
result = conn.execute("SELECT * FROM employees_offices").fetchall()

print(result)
```

```python
df = pd.DataFrame(result, columns=['employee_num','last_name','first_name','extension','email','reports_to','Title','OfficeID','city'])
df
```
